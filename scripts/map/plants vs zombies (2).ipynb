```python
cities = pd.read_excel('cities.xls')
cities.head()
```

```python
def spring(row):
    try:
        sun = Sun(row['Широта'], row['Долгота'])
        spring_sun = 0
        for dt in pd.date_range(start="2023-03-01",end="2023-05-31"):
            if (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds() > 0:
                spring_sun += (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds()
            else:
                x = (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds() + 86400
                spring_sun += x
        spring_sun /= len(pd.date_range(start="2023-03-01",end="2023-05-31"))
        spring_sun /= 3600
    except:
        spring_sun = np.nan
    return spring_sun


def summer(row):
    try:
        sun = Sun(row['Широта'], row['Долгота'])
        summer_sun = 0
        for dt in pd.date_range(start="2023-06-01",end="2023-08-31"):
            if (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds() > 0:
                summer_sun += (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds()
            else:
                x = (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds() + 86400
                summer_sun += x
        summer_sun /= len(pd.date_range(start="2023-06-01",end="2023-08-31"))
        summer_sun /= 3600
    except:
        summer_sun = np.nan
    return summer_sun

def autemn(row):
    try:
        sun = Sun(row['Широта'], row['Долгота'])
        autemn_sun = 0
        for dt in pd.date_range(start="2023-09-01",end="2023-11-30"):
            if (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds() > 0:
                autemn_sun += (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds()
            else:
                x = (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds() + 86400
                autemn_sun += x
        autemn_sun /= len(pd.date_range(start="2023-09-01",end="2023-11-30"))
        autemn_sun /= 3600
    except:
        autemn_sun = np.nan
    return autemn_sun

def winter(row):
    try:
        sun = Sun(row['Широта'], row['Долгота'])
        winter_sun = 0
        for dt in pd.date_range(start="2023-12-01",end="2024-02-29"):
            if (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds() > 0:
                winter_sun += (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds()
            else:
                x = (sun.get_local_sunset_time(dt) - sun.get_local_sunrise_time(dt)).total_seconds() + 86400
                winter_sun += x
        winter_sun /= len(pd.date_range(start="2023-12-01",end="2024-02-29"))
        winter_sun /= 3600
    except:
        winter_sun = np.nan
    return winter_sun
```

```python
cities["Весна, длина дня, ч"] = cities.apply(spring, axis = 1)
cities["Лето, длина дня, ч"] = cities.apply(summer, axis = 1)
cities["Осень, длина дня, ч"] = cities.apply(autemn, axis = 1)
cities["Зима, длина дня, ч"] = cities.apply(winter, axis = 1)
```
